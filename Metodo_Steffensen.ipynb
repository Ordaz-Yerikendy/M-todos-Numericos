# Método de Steffensen





```python
from math import *
import numpy as np
import matplotlib . pyplot as plt
```

```python
# Definimos la función
def funcion(x):
    return np.cos(x) - 2*x + 3
```

```python
# Método de Steffensen
def papa(func, p0, tol=0.00001, nmax=100):
    p = p0
    niter = 0
    error = float ('inf')

    print("# iter\t\t p \t\t f(p) \t\t error")
    print("{0} \t\t {1:6.4f} \t {2:6.4f} \t {3:6.4f}".format(niter, p, func(p), error))


    while error > tol and niter < nmax:
      # Aplicamos Método de Steffensen
      fp = func(p)
      q = p - fp
      fq = func(q)

      # Evitamos que en la división el denominador sea cero
      if fq - fp == 0:
        print ("División invalida", niter)
        break

      r =p - (fp **2) / (fq - fp)

      error = abs(r - p) # Calculamos el error
      p = r
      niter += 1
      print("{0} \t\t {1:6.6f} \t {2:6.6f} \t {3:6.6f}".format(niter, p, func(p), error))

    return p
```

```python
# Valores de entrada
p0 = 1.0 # Este es el valor que tomamos como primera aproximación
```

```python
# Ejecutamos el método
raiz = papa(funcion, p0)
print("La raíz es {0:6.7f}: ".format(raiz))
```
